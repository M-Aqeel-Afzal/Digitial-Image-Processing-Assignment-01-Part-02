```python
#Q1 a
import cv2  #importing open cv
import sys  #importing sys
import numpy as np  #importing numpy
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)
def separate_teeth(path):   #function toseparate teeths
    img1 = cv2.imread(path, cv2.IMREAD_COLOR) # reading image
    img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)  #coverting into rgb
    for i in range(0,len(img1)):  #setting red color intensity to zero of the require path of img
        for j in range(0,len(img1[0])):
            for k in range(0,len(img1[0][1])):
                if img1[i][j][k] <165:
                    img1[i][j][k]=0
    

    img1 = cv2.cvtColor(img1,cv2.COLOR_RGB2BGR)   #converting back to bgr
    cv2.imshow('final img', img1)  #dispaly img
    cv2.waitKey()
    


path = r'E:\Samester 08\Digital Image Processing\Assignments\assignment01\Assignment#1-(B)\data\dental_xray.tif' #img path

separate_teeth(path)  #function call
```

```python
#Q4 a
def count_intensity(path):
    img1 = cv2.imread(path, cv2.IMREAD_GRAYSCALE) # reading image
    counts = [0] * 256  #array initalization for counting the intensity value

    for row in img1: #reading img rows one by one
        for i in row: #reading each of row
            counts[i] += 1  # increment the count of pixals

    return counts  #returning the count
path = r'E:\Samester 08\Digital Image Processing\Assignments\assignment01\Assignment#1-(B)\data\4.grain.tif' #path
counts = count_intensity(path)  #function call

plt.hist(range(256), bins=256, weights=counts) #ploting the intensity count by histogram
plt.xlabel('Intensity')
plt.ylabel('pixals')
plt.show()
```

```python
#Q4 b
def box_filter(img,v1,v2):
    # defining the kernals
    kernel1 = [[1/49]*v1 for _ in range(7)]   #kernal 7
    kernel2 = [[1/9]*v2 for _ in range(3)]   #jernal 3

    padding1 = (len(kernel1) - 1) // 2    #size of kernal 1
    img_padded = np.zeros((img.shape[0] + padding1 * 2, img.shape[1] + padding1 * 2, 3), np.uint8)  #padding the image

    for i in range(img.shape[0]):    #padding adding
        for j in range(img.shape[1]):
            img_padded[i+padding1, j+padding1] = img[i,j]

    filter_img = np.zeros_like(img_padded)   #appling the kernal 1
    for i in range(padding1, img.shape[0] + padding1):
        for j in range(padding1, img.shape[1] + padding1):
            for k in range(3):
                filter_img[i,j,k] = np.sum(kernel1 * img_padded[i-padding1:i+padding1+1, j-padding1:j+padding1+1, k])


    filter_img = filter_img[padding1:img.shape[0]+padding1, padding1:img.shape[1]+padding1] # removing the padding
    padding2 = (len(kernel2) - 1) // 2     # size of kernal 2
    img_padded1 = np.zeros((img.shape[0] + padding2 * 2, img.shape[1] + padding2 * 2, 3), np.uint8)

    for i in range(img.shape[0]):    #padding adding
        for j in range(img.shape[1]):
            img_padded1[i+padding2, j+padding2] = img[i,j]

   
    filter_img1 = np.zeros_like(img_padded)    #appling the kernal 2
    for i in range(padding2, img.shape[0] + padding2):
        for j in range(padding2, img.shape[1] + padding2):
            for k in range(3):
                filter_img1[i,j,k] = np.sum(kernel2 * img_padded1[i-padding2:i+padding2+1, j-padding2:j+padding2+1, k])

    # Removeing padding 
    filter_img1 = filter_img1[padding2:img.shape[0]+padding2, padding2:img.shape[1]+padding2]

    for i in range(filter_img.shape[0]):   #abs negative to positive
        for j in range(filter_img.shape[1]):
            for k in range(filter_img.shape[2]):
                filter_img[i][j][k]=abs(filter_img[i][j][k]-filter_img1[i][j][k])

 
    img1_abs=cv2.cvtColor(filter_img, cv2.COLOR_BGR2GRAY)   #converting to gray scale

    for i in range(img1_abs.shape[0]):   #subtracting the original image
        for j in range(img1_abs.shape[1]):
            img[i][j]=img[i][j]-img1_abs[i][j]
    return img

# Load the image
path = r'E:\Samester 08\Digital Image Processing\Assignments\assignment01\Assignment#1-(B)\data\4.grain.tif'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
img = box_filter(img, 7,3)    
cv2.imshow('Filtered Image', img)   #display the image
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Q5
import cv2  #importing open cv
import sys  #importing sys
import numpy as np  #importing numpy
def filled_bottle(path, threshold=90 ):   #function to remove a part of an image
    img1 = cv2.imread(path, cv2.IMREAD_COLOR) # reading image
    img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)  #coverting into rgb
    count1=0
    count2=0
    for i in range(0,len(img1)):  #counting the pixals for the filled part and not filled part of the bottle
        for j in range(0,len(img1[0])):
            for k in range(0,len(img1[0][1])):
                if img1[i][j][k] <200:
                    count1+=1       #filled pivxals (black)
                else:
                    count2+=1    #not filled pixals(while)
    
   
    percentage=(count1*100)/(count1+count2)  #percentage calculating
    print("Filled percentage: {:.2f}%".format(percentage))
    if percentage < threshold:
        print("Bottle is not properly filled")
    else:
        print("Bottle is properly filled")
    img1 = cv2.cvtColor(img1,cv2.COLOR_RGB2BGR)   #converting back to bgr
    cv2.imshow('final img', img1)  #dispaly img
    cv2.waitKey()
    


path = r"E:\Samester 08\Digital Image Processing\Assignments\assignment01\Assignment#1-(B)\data\bottle2.jpg" #img path

filled_bottle(path)  #function call
```
